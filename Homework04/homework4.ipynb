## 1.  Are “deep” neural networks destined to be another bust just as perceptrons and expert systems were in the past, or are they really a breakthrough that will be used for years into the future.
In my opinion, deep neural networks are not destined to be a bust, although I do think they are currently overvalued. In the end, machine learning may well have a “dark age” in its future when people’s expectations meet reality. Both expert systems and perceptrons had their time in the sun, but eventually had their limits fully understood over time. This cycle will probably continue as people begin to learn the full limitations of deep neural networks and work is done to try to reach even further. In the future, I imagine that a new type of AI will come to fruition and become popular, perhaps even based on the ideas from deep neural networks. Deep neural networks will not become a bust per say (they have already shown they can do amazing things), but the limitations of deep neural networks will assuredly be tempered over time as more limitations are made well known.

## 3. fashion_mnist ConvNet

```python
from keras import layers
from keras import models
from keras.datasets import fashion_mnist
from keras.utils import to_categorical, plot_model
import numpy as np
import matplotlib.pyplot as plt

model = models.Sequential()

# Configure a convnet with 1 layer of convolution, max pooling, and drop out.
model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))

# Add layers to flatten the 2D image and then do a 10-way classification.
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()

# Format image data
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') / 255

#train model
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# (can add validation_split=.15 for validation set testig)
history = model.fit(train_images, train_labels, epochs=5, batch_size=128)

# check for overfitting using validation
# loss = history.history['loss']
# val_loss = history.history['val_loss']
# epochs = range(1, len(loss) + 1)
# plt.plot(epochs, loss, 'bo', label='Training loss')
# plt.plot(epochs, val_loss, 'b', label='Validation loss')
# plt.title('Training and validation loss')
# plt.xlabel('Epochs')
# plt.ylabel('Loss')
# plt.legend()
#plt.show()

#evaluate against the test data
model.evaluate(test_images, test_labels)
```

While I tested using more conv layers as well as adding dropout, the best performance
I was able to get out of my network was using a convnet with 1 layer of convolution, max pooling, and then
using a dense layer of 128 nodes. This got my accuracy to 91.24%.
